```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
dataset = pd.read_csv('heart_disease_data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 3)
```

```
print(X_train)
```

```
print(y_train)
```

```
print(X_test)
```

```
print(y_test)
```

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```
print(X_train)
```

```
print(X_test)
```

```
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 3)
classifier.fit(X_train, y_train)
```

```
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, y_train)
```

```
print('Accuracy on Training data : ', training_data_accuracy)
```

```
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, y_test)
```

```
print('Accuracy on Test data : ', test_data_accuracy)
```

```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Assuming 'sc' is your StandardScaler and 'classifier' is your trained model
# Assuming 'X_train' and 'y_train' are the training set features and labels

# Inverse transform to get the original scale of the first two features
X_set, y_set = sc.inverse_transform(X_train), y_train

# Create a mesh grid for the first two features
X1, X2 = np.meshgrid(
    np.arange(start=X_set[:, 0].min() - 10, stop=X_set[:, 0].max() + 10, step=0.25),
    np.arange(start=X_set[:, 1].min() - 10, stop=X_set[:, 1].max() + 10, step=0.25)
)

# Initialize a grid of 13 features
X_grid = np.zeros((X1.ravel().shape[0], 13))  # 13 columns for all features
X_grid[:, 0] = X1.ravel()  # Assign the first feature (X-axis)
X_grid[:, 1] = X2.ravel()  # Assign the second feature (Y-axis)

# Use sc.transform to scale the grid and predict the classifier output
Z = classifier.predict(sc.transform(X_grid)).reshape(X1.shape)

# Define colormap using distinct colors
custom_cmap = ListedColormap(['gold', 'skyblue'])

# Plot the decision boundary
plt.contourf(X1, X2, Z, alpha=0.75, cmap=custom_cmap)
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

# Plot the original training points with jitter and transparency
for i, j in enumerate(np.unique(y_set)):
    jitter = np.random.normal(0, 0.1, X_set[y_set == j, :].shape)  # Add jitter
    plt.scatter(
        X_set[y_set == j, 0] + jitter[:, 0],  # Feature 1 with jitter
        X_set[y_set == j, 1] + jitter[:, 1],  # Feature 2 with jitter
        color=['gold', 'skyblue'][i], s=50, alpha=0.5, label=f'Class {j}'
    )

# Add plot title, labels, and legend
plt.title('Logistic Regression (Training set)')
plt.xlabel('Feature 1 (Age)')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
